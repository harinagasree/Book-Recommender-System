## Colaborative Filtering - Based on user's feedback - Using Cosine Similarity

```python
books_with_ratings.head()
```

```python
x = books_with_ratings.groupby('User-ID').count()
x
```

```python
x.index
```

```python
x.shape
```

```python
# Power users - users who read more books frequently those ratings / feedback are considered
```

```python
power_users = x['Book-Rating'] > 200
power_users
```

```python
power_users = power_users[power_users].index
power_users
```

```python
power_users.shape
```

```python
power_users.sort_values()
```

```python
# selecting only the power users records from the whole table
filtered_ratings = books_with_ratings[books_with_ratings['User-ID'].isin(power_users)]
filtered_ratings
```

```python
# considering only the best users who gave atleast 200 feedback based on book-title
y = filtered_ratings.groupby('Book-Title').count()
```

```python
y
```

### The above dataframe denotes the no. of users have read the particular book

```python
y.sort_values('User-ID', ascending=False)
```

```python
y = y['User-ID'] >= 50
y
```

```python
famous_books = y[y].index
famous_books
```

```python
final_ratings = filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
final_ratings
```

```python
# Pivot table - for ratings given by each user to the each book
# book - row, user id - column
pt = final_ratings.pivot_table(index='Book-Title', columns='User-ID', values ='Book-Rating')
pt
```

```python
pt = pt.fillna(0)
pt
```

```python
similarity_scores = cosine_similarity(pt)
similarity_scores
```

```python
score_df = pd.DataFrame(similarity_scores)
score_df
```

```python
pt.index
```

```python
score_df_name = score_df
```

```python
score_df_name.index = pt.index
score_df_name.columns = pt.index
```

```python
score_df_name
```

```python
def recommend(book_name):
    index = np.where(pt.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])), key = lambda  x: x[1], reverse=True)[1:6]
    # populate the book information
    data = []
    for i in similar_items:
        item = []
        score_df_name = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(score_df_name.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(score_df_name.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(score_df_name.drop_duplicates('Book-Title')['Image-URL-M'].values))
        data.append(item)
    return data
```

```python
recommend('Animal Farm')
```
